<p>Considerando las estadísticas descriptivas, se identifica que las personas que han solicitado algun tipo de crédito estan entre los 21 y 75 años de edad, por lo tanto, se procederá a agrupar esos datos en quinquenios con el propósito de conocer donde esta la mayor concentración de créditos y en que grupo etareo.</p>

<table>
   <tr>
    <th><p>Preprocesamiento de datos con Python Pandas — Parte 5 Binning:</p></th>
    <th><p>El agrupamiento (o agrupamiento) de datos agrupa los datos en contenedores (o depósitos), en el sentido de que reemplaza los valores contenidos en un intervalo pequeño con un único valor representativo para ese intervalo. A veces, el agrupamiento mejora la precisión en los modelos predictivos.</th>
    <th><a href="https://towardsdatascience.com/data-preprocessing-with-python-pandas-part-5-binning-c5bd5fd1b950">Referencia Web</a></th>
  </tr>
</table>

```python
bins = np.linspace(21,75,12)
bins
```

```python
groupNames = ["21-25", "25-30", "30-35", "35-40", "40-45", "45-50", "50-55", "55-60", "60-65", "65-70", "70-75"]
credit['AGE_BINS'] = pd.cut(credit["AGE"], bins, labels=groupNames, include_lowest=True)
```

```python
fig,ax=plt.subplots(1,1,figsize=(16,6.9),dpi=100)

plt.hist(credit["AGE_BINS"])

#Formato del grafico
#-------Titulo y Subtitulo del grafico
title_t = "Figura 1: Cantidad de créditos"
ax.set_title(title_t,
             horizontalalignment = 'left',
             x=0.001,
             y=0.86,
             transform=fig.transFigure,
             fontsize=24,
             fontweight="bold",
             fontfamily="calibri",c="#F352D5")

subtitle_t = 'Créditos otorgados segun grupo etario.'
plt.suptitle(subtitle_t,
            horizontalalignment = "left",
            x = 0.001,
            y = 0.84,
            transform = fig.transFigure,
            fontsize=16,
            fontweight="regular",
            fontfamily="calibri", c="purple")

#-------Etiquetas en Eje X y Y
label = "Grupos Etarios"
ax.set_xlabel(label,
             ha="left",
             fontsize=10,
             fontweight="regular",
             fontfamily="calibri", c="purple")

ax.xaxis.set_label_coords(x=0,
                          y=-0.07,
                          transform=ax.transAxes)

ax.tick_params(axis='both',
               labelsize=12)

plt.subplots_adjust(top=0.70)

plt.show()
```
